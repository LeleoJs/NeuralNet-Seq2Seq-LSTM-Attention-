```python
import numpy as np
import matplotlib.pyplot as plt
import ewtpy

# Parmetros do sinal
T = 1000
t = np.arange(1, T+1) / T

# Sinal original
f = np.cos(2 * np.pi * 0.8 * t) + 2 * np.cos(2 * np.pi * 10 * t) + 0.8 * np.cos(2 * np.pi * 100 * t)

# Aplicar a EWT
ewt, mfb, boundaries = ewtpy.EWT1D(f, N=4)

# Reconstruir o sinal a partir dos modos empricos
reconstructed_signal = np.sum(ewt, axis=1)  # Soma ao longo do eixo 1 (modo)

# Plotar os resultados
plt.figure(figsize=(12, 6))

# Plotar o sinal original
plt.subplot(2, 1, 1)
plt.plot(t, f, label='Sinal Original', color='blue')
plt.title('Sinal Original')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.legend()

# Plotar o sinal reconstrudo
plt.subplot(2, 1, 2)
plt.plot(t, reconstructed_signal, label='Sinal Reconstrudo', color='orange')
plt.title('Sinal Reconstrudo a partir da EWT')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```
